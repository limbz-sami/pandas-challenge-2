```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete.head()
```

```python
Total_school = len(school_data_complete['school_name'].unique())
Total_student = school_data_complete['Student ID'].count()
School_unique_data = school_data_complete.drop_duplicates('school_name')
Total_budget = School_unique_data['budget'].sum()
Avg_math = school_data_complete['math_score'].mean()
Avg_reading = school_data_complete['reading_score'].mean()
Math_pass_data = school_data_complete.loc[school_data_complete['math_score'] >= 70,:]
passing_math = Math_pass_data['math_score'].count()/len(school_data_complete['math_score'])*100
Reading_pass_data = school_data_complete.loc[school_data_complete['reading_score'] >= 70,:]
passing_reading = Reading_pass_data['reading_score'].count()/len(school_data_complete['reading_score'])*100
Overall_passing = (passing_math + passing_reading)/2


Summary_table_1 = pd.DataFrame({
                    "Total Schools": [Total_school],
                    "Total Students": [Total_student],
                    "Total Budget": [Total_budget],
                    "Average Math Score": [Avg_math],
                    "Average Reading Score": [Avg_reading],
                    "% Passing Math": [passing_math],
                    "% Passing Reading": [passing_reading],
                    "% Overall Passing Rate": [Overall_passing]})
Summary_table_1['Total Budget']=Summary_table_1['Total Budget'].map('${:,}'.format)
Summary_table_1['Average Math Score']=Summary_table_1['Average Math Score'].map('{:.2f}%'.format)
Summary_table_1['Average Reading Score']=Summary_table_1['Average Reading Score'].map('{:.2f}%'.format)
Summary_table_1['% Passing Math']=Summary_table_1['% Passing Math'].map('{:.2f}%'.format)
Summary_table_1['% Passing Reading']=Summary_table_1['% Passing Reading'].map('{:.2f}%'.format)
Summary_table_1['% Overall Passing Rate']=Summary_table_1['% Overall Passing Rate'].map('{:.2f}%'.format)
Summary_table_1
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
group_school = school_data_complete.groupby('school_name')
Total_student = group_school['Student ID'].count()
Total_School_budget = group_school['budget'].mean()
Per_student_budget = Total_School_budget/Total_student
Avg_math_score = group_school['math_score'].mean()
Avg_reading_score = group_school['reading_score'].mean()

math_pass_data = school_data_complete.loc[(school_data_complete['math_score']>=70),:]
math_grouped = math_pass_data.groupby("school_name")
passing_math = ((math_grouped['math_score'].count())/Total_student)*100
reading_pass_data = school_data_complete.loc[(school_data_complete['reading_score']>=70),:]
reading_grouped = reading_pass_data.groupby('school_name')
passing_reading = ((reading_grouped['reading_score'].count())/Total_student)*100
overall_pass_rate = (passing_math + passing_reading)/2

Summary_table_1 = pd.DataFrame({
                    "Total Student": Total_student,
                    "Total School Budget": Total_School_budget,
                    "Per Student Budget":Per_student_budget,
                    "Average Math Score":Avg_math_score,
                    "Average Reading Score": Avg_reading_score,
                    "% Passing Math": passing_math,
                    "% Passing Reading": passing_reading,
                    "% Overall Passing Rate": overall_pass_rate
                    })

Sorted_1 = Summary_table_1.sort_values('% Overall Passing Rate', ascending= False)
Sorted_1['Total School Budget'] = Sorted_1['Total School Budget'].map("${:,}".format)
Sorted_1['Per Student Budget'] = Sorted_1['Per Student Budget'].map("${:,}".format)
Sorted_1['Average Math Score'] = Sorted_1['Average Math Score'].map("${:.2f}%".format)
Sorted_1['Average Reading Score'] = Sorted_1['Average Reading Score'].map("${:.2f}%".format)
Sorted_1['% Passing Math'] = Sorted_1['% Passing Math'].map("${:.2f}%".format)
Sorted_1['% Passing Reading'] = Sorted_1['% Passing Reading'].map("${:.2f}%".format)
Sorted_1['% Overall Passing Rate'] = Sorted_1['% Overall Passing Rate'].map("${:.2f}%".format)
Sorted_1.head(5)

```
